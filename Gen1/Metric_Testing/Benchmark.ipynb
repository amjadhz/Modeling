```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('../csv/All_Demand.csv', index_col=0)
df = df.drop(['Unnamed: 0.1'], axis=1)
df['Record Date'] = pd.to_datetime(df['Record Date'])
df = df.drop(index=1825)
df = df.sort_values('Record Date')
df['Air Mean'] = np.mean(df["Air Temperature"])
df['Demand Mean'] = np.mean(df["Total Demand"])
```

```python
fig, ax1 = plt.subplots()

ax1.plot(df["Record Date"], df["Total Demand"], color = "red")
ax1.plot(df["Record Date"], df["Demand Mean"], color = "blue")

ax2 = ax1.twinx()

plt.figure(figsize=(10, 10))
plt.show()
```
